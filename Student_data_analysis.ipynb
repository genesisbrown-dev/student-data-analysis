```python
import pandas as pd
# Import numpy for numerical operations
import numpy as np
# Import matplotlib for creating graphs
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame (table)
# Make sure your file name matches exactly
Student_Performance = pd.read_csv("Student_Performance.csv")

# Show first 5 rows
Student_Performance.head()
```

```python
# Understanding the dataset:
# Show general information about the dataset
# Includes column names, data types, and missing values(Non -null values)
print("Dataset Info:")
Student_Performance.info()
```

```python
# .describe() Show statistical summary of numerical columns
# Includes mean (average), min, max, etc.
print("\nStatistical Summary:")
Student_Performance.describe()
```

Column Names

```python
## Column names:
# Display all column names in the dataset
# Useful to know what data you can analyze
print("Columns in dataset:")
print(Student_Performance.columns)
```

```python
# Calculate average (mean) scores
print("Average Scores:")

# .mean() calculates the average value of a column
print("Math:", Student_Performance['math_score'].mean())
print("Reading:", Student_Performance['science_score'].mean())
print("Writing:", Student_Performance['english_score'].mean())

# Find highest values in each column
print("\nHighest Scores:")

# .max() returns the highest value in a column
print("Math:", Student_Performance['math_score'].max())
print("Reading:", Student_Performance['science_score'].max())
print("Writing:", Student_Performance['english_score'].max())
```

```python
# Group data by 'gender' and calculate average scores
# This helps compare performance between groups

grouped = Student_Performance.groupby('gender')[['math_score', 'english_score', 'science_score']].mean()

print("Average Scores by Gender:")
print(grouped)
```

# Histogram (distribution) -Visualization


```python
# Create a histogram for math scores
# Shows how scores are distributed

Student_Performance['math_score'].hist()

# Add title and labels to make graph readable
plt.title("Distribution of Math Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")

# Display the graph
plt.show()
```

## Bar Chart (comparison) Visualization

```python

# Create a bar chart from grouped data
# This compares average scores between genders

grouped.plot(kind='bar')

# Add title and label
plt.title("Average Scores by Gender")
plt.ylabel("Score")

# Show the graph
plt.show()
```
